```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("/content/aqi_data/aqi.csv")
df.columns = df.columns.str.strip()
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["aqi_value"] = pd.to_numeric(df["aqi_value"], errors="coerce")

df = df.dropna(subset=["date", "aqi_value"])
df = df.reset_index(drop=True)

df.info()
```

```python
df_bhopal = df[(df["state"] == "Madhya Pradesh") & (df["area"].str.contains("Bhopal", case=False, na=False))].copy()
df_bhopal = df_bhopal.sort_values("date").reset_index(drop=True)

print("Rows:", len(df_bhopal))
df_bhopal.head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df_bhopal["date"], df_bhopal["aqi_value"])
plt.title("Bhopal AQI Trend (Raw)")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_bhopal["aqi_rolling_30"] = df_bhopal["aqi_value"].rolling(30).mean()

plt.figure(figsize=(14,5))
plt.plot(df_bhopal["date"], df_bhopal["aqi_value"], alpha=0.4, label="Raw AQI")
plt.plot(df_bhopal["date"], df_bhopal["aqi_rolling_30"], linewidth=2, label="30-day Rolling Mean")
plt.title("Bhopal AQI Trend (Smoothed)")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df_feat = df_bhopal[["date", "aqi_value", "number_of_monitoring_stations"]].copy()

df_feat["number_of_monitoring_stations"] = pd.to_numeric(df_feat["number_of_monitoring_stations"], errors="coerce")
df_feat["number_of_monitoring_stations"] = df_feat["number_of_monitoring_stations"].fillna(df_feat["number_of_monitoring_stations"].median())

for lag in [1, 2, 3, 7, 14]:
    df_feat[f"lag_{lag}"] = df_feat["aqi_value"].shift(lag)

df_feat["rolling_3"] = df_feat["aqi_value"].shift(1).rolling(3).mean()
df_feat["rolling_7"] = df_feat["aqi_value"].shift(1).rolling(7).mean()
df_feat["rolling_14"] = df_feat["aqi_value"].shift(1).rolling(14).mean()

df_feat["year"] = df_feat["date"].dt.year
df_feat["month"] = df_feat["date"].dt.month
df_feat["day"] = df_feat["date"].dt.day
df_feat["dayofweek"] = df_feat["date"].dt.dayofweek

df_feat = df_feat.dropna().reset_index(drop=True)

df_feat.head()
```

```python
X = df_feat.drop(columns=["date", "aqi_value"])
y = df_feat["aqi_value"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

rf = RandomForestRegressor(n_estimators=300, random_state=42)
rf.fit(X_train, y_train)

print("Models trained successfully ✅")
```

```python
def evaluate_model(name, model, X_test, y_test):
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    print(name)
    print("MAE  :", round(mae, 2))
    print("RMSE :", round(rmse, 2))
    print("R2   :", round(r2, 3))
    print("-" * 30)

evaluate_model("Linear Regression", lr, X_test, y_test)
evaluate_model("Random Forest", rf, X_test, y_test)
```
